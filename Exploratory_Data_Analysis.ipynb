```python
!pip install gdown

import gdown
import pandas as pd
from sklearn.preprocessing import StandardScaler

url = "https://drive.google.com/uc?id=1yLbilHV68w0c5ByHzrw0ohcTATDZkeY4"
output = "toyota_corolla.csv"
gdown.download(url, output, quiet=False)

df = pd.read_csv(output)

print(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for numerical features
df.hist(bins=30, figsize=(15, 10))
plt.show()

# Box plot for numerical features
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.show()

# Scatter plot for feature relationships
sns.pairplot(df)
plt.show()

from sklearn.preprocessing import StandardScaler

numerical_features = ['Price', 'Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears', 'Quarterly_Tax', 'Weight'] # Replace with your desired numerical features

# Normalizing numerical features
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```
